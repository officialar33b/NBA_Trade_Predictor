```python
def was_traded(player_df, season_list):
    # Sort seasons chronologically
    sorted_seasons = sorted(season_list)
    player_df = player_df.sort_values('SEASON_ID')
    
    # Initialize column to track trades
    player_df['CHANGED_FRANCHISE'] = False
    
    for i in range(1, len(sorted_seasons)):  # Start from 1 to avoid index -1
        current_season = sorted_seasons[i]
        prev_season = sorted_seasons[i-1]
        
        # Check if player participated in both seasons
        current_in_df = current_season in player_df['SEASON_ID'].values
        prev_in_df = prev_season in player_df['SEASON_ID'].values
        
        if current_in_df and prev_in_df:
            current_team = player_df[player_df['SEASON_ID'] == current_season]['TEAM_ID'].iloc[0]
            prev_team = player_df[player_df['SEASON_ID'] == prev_season]['TEAM_ID'].iloc[0]
            
            if current_team != prev_team:
                player_df.loc[player_df['SEASON_ID'] == current_season, 'CHANGED_FRANCHISE'] = True
                
        elif current_in_df and not prev_in_df:
            # Handle rookies (no prior season)
            player_df.loc[player_df['SEASON_ID'] == current_season, 'CHANGED_FRANCHISE'] = False
    
    return player_df
```
