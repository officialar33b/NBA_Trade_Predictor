### EFF of 13 or greater.

Current Players

```python
per_leaders = leagueleaders.LeagueLeaders(stat_category_abbreviation='EFF').get_data_frames()[0]
per_leaders.head(10)
```

```python
per_leaders['EFF_PER_GAME'] = per_leaders['EFF']/per_leaders['GP'] # EFF / games played
per_leaders.head(10)
```

```python
high_per_players = per_leaders[per_leaders['EFF_PER_GAME'] >= 13]
high_per_players.head(10)
```

```python
high_per_players.tail(10)
```

```python
high_per_players.to_csv(r'./current_superstars_2024_25.csv')
```
